## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') #strategy='constant', fill_value=5
X[:, 1:2] = imputer.fit_transform(X[:, 1:2])#second col er missing val change korlam median diye. chaile 1:3 kore 2nd,3rd duto col er miss val e median/mean diye change korte partam
#X = imputer.fit_transform(X) eta eikhetre use korte parbo na karon X er prothom coloumn a non-numeric number ache,tai shuhdhu numeric col gulo sleect kore dite hobe ie khetre
imputer = SimpleImputer(missing_values=np.nan, strategy='median') #strategy='constant', fill_value=5
X[:, 2:3] =imputer.fit_transform(X[:, 2:3]) #third col er missing val chnage korlam constant diye.
```
