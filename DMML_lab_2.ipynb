```python
import pandas as pd
import numpy as np
```

**Pandas Basic**

```python
df=pd.DataFrame(
    {
        "Name" : ["Tawfek","Negar","Apra","Tuhin","Mahtab","Khalid",],
        "Age" : [22,20,21,22,23,22],
        "City" : ["Bhola","Dhaka","Sylhet","Madaripur",np.nan,"Laxmipur"]
    }
)
print(df)
```

## **Viewing Data**

```python
print(df.info())
print(df.shape)
print(df.head(2))
print(df.tail())
```

**Selecting Dta**

```python
print(df["Name"])  # .Age means dot notation of dataframe df
print(df["Age"])  #Safer way and select single row
print(df[["Name","Age"]])   #Select multiple cols [[]] means

# print(df[1:2,0:1])   this is actually wrong
# print(df['Age':'City'])   this is actually wrong
```

**Loc and iloc**

```python
# loc --> label
# iloc --> index location

print(df.loc[[2,1]])
print(df.loc[1:2,'Name':'City'])  #range of column
print(df.loc[[1,3],['Name','City']])  #seperate rows and columns
# print([df.loc[1:2,1:2]])  this is wrong
print(df.iloc[:,0:2])
print(df.iloc[[2,5],[0,2]])
```

**Filtering rows and column**

```python
print(df['Age']>21)
print(df.loc[df["Age"]>21])
```
